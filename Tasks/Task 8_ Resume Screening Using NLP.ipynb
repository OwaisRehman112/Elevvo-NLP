```python
import os
import pandas as pd
import kagglehub

# Load the datasets
path1 = kagglehub.dataset_download("snehaanbhawal/resume-dataset")
path2 = kagglehub.dataset_download("ravindrasinghrana/job-description-dataset")

# List the files in dataset directory
print("files in dataset: ", os.listdir(path1))
print("files in dataset: ", os.listdir(path2))


resume_path = os.path.join(path1, 'Resume', 'Resume.csv')
job_path = os.path.join(path2, 'job_descriptions.csv')


resume_df = pd.read_csv(resume_path)
job_df = pd.read_csv(job_path)


resume_df.head(15)
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('all-MiniLM-L6-v2')

def screen_and_rank_resumes(job_index, num_resumes=50):
    job_desc = job_df.iloc[job_index]['Job Description']
    job_title = job_df.iloc[job_index]['Job Title']

    resumes_to_check = resume_df.head(num_resumes).copy()
    resume_texts = (resumes_to_check['Category'] + " " + resumes_to_check['Resume_str']).tolist()

    # Generate embeddings
    job_vector = model.encode(job_desc, convert_to_tensor=True)
    resume_vectors = model.encode(resume_texts, convert_to_tensor=True)

    # Calculate ranking scores
    scores = util.cos_sim(job_vector, resume_vectors)[0]
    resumes_to_check['Match_Score'] = scores.tolist()

    return job_title, resumes_to_check.sort_values(by='Match_Score', ascending=False)

# Run system
title, ranked_resumes = screen_and_rank_resumes(0)
print(f"Top Matches for: {title}")
print(ranked_resumes[['ID', 'Category', 'Match_Score']].head(10))
```

```python
# Use a smaller sample to save time
sample_size = 10000
print(job_df.shape)

job_list_small = job_df['Job Description'].head(sample_size).tolist()
resume_list_small = (resume_df['Category'] + " " + resume_df['Resume_str']).head(sample_size).tolist()

# Encode with progress bar
job_embeddings = model.encode(job_list_small, convert_to_tensor=True, show_progress_bar=True)
resume_embeddings = model.encode(resume_list_small, convert_to_tensor=True, show_progress_bar=True)
```

```python
from sentence_transformers import util

# Calculate similarity scores
# Comparing the first job embedding against all resume embeddings
cosine_scores = util.cos_sim(job_embeddings[0], resume_embeddings)[0]

# Create a results dataframe for the sampled resumes
results_df = resume_df.head(len(resume_list_small)).copy()
results_df['Match_Score'] = cosine_scores.tolist()

# Rank resumes by score
ranked_resumes = results_df.sort_values(by='Match_Score', ascending=False)

# Display results
print(f"Target Job: {job_df.iloc[0]['Job Title']}")
print(ranked_resumes[['ID', 'Category', 'Match_Score']].head(10))
```

```python
import spacy

# Load the model
nlp = spacy.load("en_core_web_sm")

def extract_resume_info(text):
    doc = nlp(text)

    # Extracting entities
    # ORG: Companies/Schools, DATE: Tenure, PRODUCT/GPE: Tech/Locations
    entities = {
        "Organizations": list(set([ent.text for ent in doc.ents if ent.label_ == "ORG"])),
        "Experience_Timeline": list(set([ent.text for ent in doc.ents if ent.label_ == "DATE"]))
    }
    return entities

# Apply to your top 5 ranked resumes
ranked_resumes['Extracted_Info'] = ranked_resumes['Resume_str'].head(5).apply(extract_resume_info)

# Display a sample
import json
print(json.dumps(ranked_resumes.iloc[0]['Extracted_Info'], indent=2))
```

```python
import torch

# Save the math (embeddings)
torch.save(resume_embeddings, 'resume_embeddings.pt')

# Save the metadata (so we can map vectors back to names/IDs)
# We only need the ID and Category to keep the file small
resume_df[['ID', 'Category', 'Resume_str']].head(sample_size).to_csv('resume_metadata.csv', index=False)
```
